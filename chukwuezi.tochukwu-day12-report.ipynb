### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.



````python
#set figure size
plt.figure(figsize=(10,6))

#define the function [data] to plot
x = np.linspace(0, 10, 100)
y = x

#the actual plotting
plt.plot(x, y, linewidth=2, label=r'$y = x$')

# x and y ticks
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(0, 11, 1))

# x and y limits
plt.xlim(0, 10)
plt.ylim(0, 10)

# Labels
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')

# Title of the plot
plt.title('Simple Line Plot')

# show the plot
plt.show()
````

### Exercise 2:
Plot `y = x^2` using Matplotlib for `x` values between -10 and 10.



````python
# Set the figure size
plt.figure(figsize=(10,6))

# Define the function [data] to plot
x = np.linspace(-10, 10, 100)
y = x**2

# The actual plotting
plt.plot(x, y, color='red', label=r'$y = x^2$', linewidth=2)

# X and y ticks

plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(0, 101, 10))

# X and y limits
plt.xlim(-10, 10)
plt.ylim(0, 100)

# Labels

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')

# Title of the plot
plt.title(r'Graph of $y=x^2$')

# Show the plot

plt.show()
````

### Exercise 4:
Plot a sine wave using Matplotlib for `x` values between 0 and 2.


````python
# Set up the figure size

plt.figure(figsize=(20,6))

# Generate random data

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Plot the line plot

plt.plot(x, y, color='blue', label=r'$y = \sin(x)$')

# X and y ticks

plt.xticks(np.arange(0, 2*np.pi + np.pi/4, np.pi/4), [0, r'$+\pi/4$', r'$+\pi/2$', r'$+3\pi/4$', r'$+\pi$', r'$+5\pi/4$', r'$+3\pi/2$',r'$+7\pi/4$',r'$+2\pi$'])
plt.yticks(np.arange(-1, 1.1, 0.2))

# X and y limits

plt.xlim(0, 2*np.pi)
plt.ylim(-1.1, 1.1)

# Grids

plt.grid(True)

# Labels

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')

# Title of the plot

plt.title(r'Graph of $\sin(x)$')

# Show the plot

plt.show()
````

### Exercise 6:

### Exercise 8:
Add titles, axis labels, and a legend to the plot in Exercise 7.


````python
# Create the figure
plt.figure(figsize=(15,6))

# Generate the data to plot
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x**3

# The plot
plt.plot(x, y1, label=r"$y = x^2$")
plt.plot(x, y2, label=r"$y = x^3$")
plt.legend(loc="best")
plt.grid()

# Axis labels
plt.xlabel("x")
plt.ylabel(r"$y_1 = x^2$ and $y_2 = x^3$")

plt.show()
````

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

````python
# Create the figure
fig, axes = plt.subplots(2,1, figsize=(15,15))

# Generate the data to plot
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x**3

# The plot
axes[0].plot(x, y1, c="red", label=r"$y = x^2$")
axes[1].plot(x, y2, label=r"$y = x^3$")
axes[0].legend(loc="best")
axes[1].legend(loc="best")
axes[0].grid()
axes[1].grid()

# Axis labels

axes[0].set_xlabel(r"$x$")
axes[0].set_ylabel(r"$y_1 = x^2$")
axes[1].set_xlabel(r"$x$")
axes[1].set_ylabel(r"$y_2 = x^3$")

# help(fig)
fig.legend(loc="upper right")
fig.tight_layout()

plt.show()
````

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

````python
# Set up the figure

plt.figure(figsize=(15,10))

# Generate the data to plot

x = np.linspace(0, 50, 500)
y1 = x
y2 = x**2
y3 = x**3

# The plot

plt.plot(x, y1, c="blue", linewidth=2, label=r"$y = x$", ls="--")
plt.plot(x, y2, c="green", lw =2, label=r"$y = x^2$", ls="-.")
plt.plot(x, y3, c="red", lw=2, label=r"$y = x^3$", ls=":")

plt.legend(loc="best")
plt.grid()

# X and Y limits
plt.xlim(0, 50)
plt.ylim(0, 50)

# Lablels
plt.xlabel("x")
plt.ylabel("y")
plt.title("Three in one")

# Show the plot
plt.show()
````

### Exercise 14:
Plot a filled area under the curve of `y = x^2` for `x` ranging from 0 to 10.

````python
# Generate the figure

plt.figure(figsize=(10,6))

# Generate the data to plot
x = np.linspace(0, 10, 100)
y = x**2

# Plot the filled area
plt.plot(x,y, lw=3, ls=':', label=r"$y = x^2")
plt.fill_between(x, -0, y, color='orange', alpha=0.5)


# Set the axis with grids and labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Filled Area Plot')

# Show the plot

plt.show()
````

### Exercise 15:
Change the style of the plot to `ggplot` and replot the sine wave from Exercise 4.


````python
# Set the figure size
plt.figure(figsize=(20,6))


# Generate the data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Use the ggplot style to plot
plt.style.use('ggplot')
plt.plot(x, y, label='Sine Wave')

# X and y ticks
plt.xticks(np.arange(0, 2*np.pi + np.pi/4, np.pi/4), [0, r'$+\pi/4$', r'$+\pi/2$', r'$+3\pi/4$', r'$+\pi$', r'$+5\pi/4$', r'$+3\pi/2$',r'$+7\pi/4$',r'$+2\pi$'])
plt.yticks(np.arange(-1, 1.1, 0.2))

# X and y limits
plt.xlim(0, 2*np.pi)
plt.ylim(-1.1, 1.1)

# Grids
plt.grid(True)

# Labels
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')

# Title of the plot
plt.title(r'Graph of $\sin(x)$ with ggplot style')

# Show the plot

plt.show()

````

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.

````python
# Set the figure size

plt.figure(figsize=(10,6))

# Define the categories and values

categories = ['X', 'Y', 'Z']
values_A = [3, 5, 7]
values_B = [2, 6, 4]

# Create the stacked bar chart

plt.bar(categories, values_A, label='Group A', color='darkblue', width=0.4, alpha=0.8)
plt.bar(categories, values_B, label='Group B', color='darkgreen', width=0.4, bottom=values_A, alpha=0.7)

# Set the axis with grids and labels

plt.grid(True)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')

# Show the legend

plt.legend(loc='best')

# Show the plot

plt.show()
````

### Exercise 19:
#Add error bars to a bar plot with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`, assuming the error is 1 for each value.


````python
# Set the figure size

plt.figure(figsize=(10,6))

# Define the categories and values

categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

# Plot the bar chart

plt.bar(categories, values, label='Values', color='darkblue', alpha=0.7)

# Add error bars to the bar chart

plt.errorbar(categories, values, yerr=1, fmt='o', color='red')

# Set the axis with grids and labels

plt.grid(True)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Error Bars')

# Show the legend

plt.legend(loc='best')

# Show the plot

plt.show()
````

### Exercise 20:
Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.



````python
# Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.

# Create the figure
plt.figure(figsize=(20, 6))

# Generate the data to plot
x = np.linspace(0, 6*np.pi + np.pi/2, 100)

# Plot the sine and cosine waves
plt.plot(x, np.sin(x), 'r--', label=r'$\sin(x)$')
plt.plot(x, np.cos(x), 'b-', label=r'$\cos(x)$')

# Set the x and y ticks

plt.xticks(np.arange(0, 6*np.pi + np.pi/2, np.pi/2))
plt.yticks(np.arange(-1, 1.2, 0.2))

# Set the x and y limits

plt.xlim(0, 6*np.pi+np.pi/2)
plt.ylim(-1.5, 1.5)


# Add labels and legend
plt.legend(loc='best')
plt.grid()

# Show the plot
plt.show()
````

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.
